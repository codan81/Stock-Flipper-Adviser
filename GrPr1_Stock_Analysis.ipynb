#### Imputing data from Alpaca

```python
import pandas as pd
import os
from dotenv import load_dotenv
import alpaca_trade_api as tradeapi
from MCForecastTools import MCSimulation
from datetime import datetime
```

```python
load_dotenv()
```

```python
# Setting the variables for the Alpaca API and secret keys
alpaca_api_key = os.getenv('APCA_API_KEY_ID')
alpaca_api_secret_key = os.getenv('APCA_API_SECRET_KEY')

display(type(alpaca_api_key))
display(type(alpaca_api_secret_key))

# Create the Alpaca tradeapi.REST object
alpaca_api = tradeapi.REST(alpaca_api_key, alpaca_api_secret_key, api_version = "v2")
```

```python
# Set start and end dates of 10 years back from your current date
timeframe="1D"
ten_years_ago = pd.Timestamp.now(tz="America/New_York").year-10
today=pd.Timestamp.now()
start_date= pd.Timestamp(today- pd.Timedelta(days=3650), tz="America/New_York").isoformat()
end_date = pd.Timestamp(today).isoformat()
# Set number of rows to 1000 to retrieve the maximum amount of rows
max_rows=1000
```

```python

"""
Shopify (NYSE:SHOP)
Square (NYSE:SQ)
Chipotle (NYSE:CMG)
Ross (NASDAQ:ROST)
Crocs (NASDAQ:CROX)
Amazon (NASDAQ:AMZN)
BioCryst Pharma (NASDAQ:BCRX)
Moderna (NasdaqGS: MRNA)
"""

tickers=["SHOP","SQ","CMG","ROST","CROX","AMZN","BCRX","MRNA"]
```

```python
historical_df=alpaca_api.get_barset(
    tickers,
    timeframe,
    start=start_date,
    end=end_date,
    limit=max_rows
    ).df

# Display both the first and last five rows of the DataFrame
historical_df.head()
historical_df.tail()
```
